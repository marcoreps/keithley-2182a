```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
from matplotlib.ticker import EngFormatter
from matplotlib.ticker import FormatStrFormatter

# Array from .csv
arr=np.genfromtxt('C:/Users/reps/Documents/keithley-2182a/20240425-231708_Keithley_2182a_short_NPLC5.csv',delimiter=',')
f, S = scipy.signal.welch(arr, 10, nperseg=65000)
f = f[1:-1]
S = np.sqrt(S[1:-1])


fig,ax = plt.subplots(figsize=(16,9))

#plt.figure(facecolor='white') # fix for exporting png file
plt.loglog()
plt.xlim([0.0001,1])
plt.xlabel('frequency [Hz]')
ax.xaxis.set_major_formatter(FormatStrFormatter('%.4f'))
ax.yaxis.set_major_formatter(EngFormatter())
ax.yaxis.set_minor_formatter(EngFormatter())

#plt.yticks((1E-9, 3E-9, 1E-8, ))
plt.ylabel('PSD [V/rHz]')
plt.grid(True, which="both")
plt.title('Shorted Voltmeters')
plt.plot(f, S, label='2182A SYST:AZER:STAT ON assuming 5 NPLC', alpha=0.7)


arr=np.genfromtxt('C:/Users/reps/Documents/keithley-2182a/20240425-231708_Keithley_2182a_short_NPLC5.csv',delimiter=',')
f, S = scipy.signal.welch(arr, 2.143, nperseg=18000)
f = f[1:-1]
S = np.sqrt(S[1:-1])
plt.plot(f, S, label='2182A SYST:AZER:STAT ON assuming reading output rate', alpha=0.7)



arr=np.genfromtxt('C:/Users/reps/Documents/keithley-2182a/20240419-154946_Keithley_2182a_short_NPLC5.csv',delimiter=',')
f, S = scipy.signal.welch(arr, 10, nperseg=25000)
f = f[1:-1]
S = np.sqrt(S[1:-1])
plt.plot(f, S, label='2182A SYST:AZER:STAT OFF assuming 5 NPLC', alpha=0.7)

arr=np.genfromtxt('C:/Users/reps/Documents/keithley-2182a/20240425-003650_Keithley_2182a_short_NPLC5.csv',delimiter=',')
f, S = scipy.signal.welch(arr, 4, nperseg=25000)
f = f[1:-1]
S = np.sqrt(S[1:-1])
plt.plot(f, S, label='2182A SYST:AZER:STAT OFF assuming reading output rate', alpha=0.7)


plt.legend()
```
